```python
import cv2
```

```python
a_count = 0
b_count = 0
```

```python
cam = cv2.VideoCapture(0) # creates cam object for your webcam
key = '' # initializes key variable to contain no character.
while key != 113: # checks if key equals 113 which is ascii for q
    ret, frame = cam.read() # gets image from camera and stores it in frame
    frame = cv2.resize(frame, (240, 240)) # resizes the image to 240 x 240
    cv2.imshow("hello", frame) # shows the image in a named window
    key = cv2.waitKey(10) # waits 10 ms for keyboard input and stores the input in key, blank if no key pressed
    if key == 97: # checks if key variable equals 97 which is ascii for a
        cv2.imwrite("Dataset/a/a_" + str(a_count) + ".png", frame) # saves image by combining a fixed path with the count
        print("saved a_" + str(a_count)) # prints the number of image saved to help keep count
        a_count += 1 # adds 1 to count
    if key == 98: # same as above but for b
        cv2.imwrite("Dataset/b/b_" + str(b_count) + ".png", frame)
        print("saved b_" + str(b_count))
        b_count += 1
cam.release() # releases the webcam device
cv2.destroyAllWindows() # destroys the named window
```
